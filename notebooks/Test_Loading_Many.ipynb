```python
from pystac import Item
from odc.stac import load
from odc.geo.geom import BoundingBox
from rustac import read_sync
from obstore.store import S3Store
```

```python
parquet = "s3://data.ldn.auspatious.com/ausp_ls_geomad/0-0-2/ausp_ls_geomad.parquet"
stac_docs = read_sync(parquet, store=S3Store(bucket="data.ldn.auspatious.com", region="us-west-2"))
print(f"Found {len(stac_docs["features"])} STAC documents")
items = [Item.from_dict(feature) for feature in stac_docs["features"]]
print(f"Parsed {len(items)} STAC items")
```

```python
# BBOX over Singapore
# bbox = BoundingBox(left=103.60, bottom=1.20, right=104.05, top=1.55)

# BBOX over SW Viti Levu in Fiji
bbox = BoundingBox(left=177.4, bottom=-18.3, right=178.0, top=-17.8)

bbox.explore()
```

```python
data = load(items, bbox=bbox, measurements=["red", "green", "blue"], chunks={})
data
```

```python
# Plot an RGB image for every 5 years of data
data.sel(time=["2000", "2005", "2010", "2015", "2020", "2025"]).to_array().plot.imshow(col="time", col_wrap=2, size=6, robust=True)
```
